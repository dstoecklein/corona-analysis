```python
import pandas as pd
pd.set_option("display.max_columns", None)
import datetime as dt
from utils import db_helper as database
```

```python
db = database.DB()
```

```python
df = db.get_table("covid_daily")
df_calendar_days = db.get_table("_calendar_days")
```

```python
df.columns
```

```python
df_calendar_days.columns
```

```python
tmp = df.merge(df_calendar_days,
                left_on='calendar_days_fk',
                right_on='calendar_days_id',
                how='left',
                )
```

```python
tmp.head()
```

```python
if 'iso_day' in tmp.columns:
        try:
            tmp['iso_day'] = pd.to_datetime(
                tmp['iso_day'], infer_datetime_format=True
            ).dt.date
            tmp['iso_day'] = pd.to_datetime(
                tmp['iso_day'], infer_datetime_format=True
            )
        except (KeyError, TypeError):
            print("Error trying to convert Date columns")
```

```python
df_2020 = tmp[tmp['iso_day']==dt.datetime(2020,12,31)]
df_2021 = tmp[tmp['iso_day']==dt.datetime(2021,12,31)]
```

```python
dt.date.today().day
```

```python
df_2022 = tmp[tmp['iso_day']==dt.datetime(dt.date.today().year,dt.date.today().month,dt.date.today().day)]
```

```python
df_2022.head()
```

```python
df_2020.head()
```

```python
df_2021.head()
```

```python
test = pd.concat([df_2020, df_2021, df_2022])
test.set_index(test['iso_day'].dt.year, inplace=True)
```

```python
test.head()
```

```python
test.columns
```

```python
test = test._get_numeric_data()
```

```python
test.head()
```

```python
test.loc[2021] = test.loc[2021] - test.loc[2020]
test.loc[2022] = test.loc[2022] - (test.loc[2021] + test.loc[2020])
```

```python
test.head()
```

```python
test.reset_index(inplace=True)
test.rename(columns={'iso_day': 'iso_year'}, inplace=True)
```
