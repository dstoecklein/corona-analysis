```python
import pandas as pd
from pygenesis.py_genesis_client import PyGenesisClient
```

```python
username = 'XXX'
password = 'XXX'
client = PyGenesisClient(
        site='DESTATIS',
        username=username,
        password=password
    )

df = client.read('12411-0015')
```

```python
df.head()
```

```python
df2 = df.melt(
        id_vars=["index.0", "index.1"], var_name="stichtag", value_name="population"
)
```

```python
df2.rename(columns={
    'index.0': 'ags',
    'index.1': 'subdivision_3',
    'stichtag': 'year',
    'population': 'population'
}, inplace=True)
```

```python
df2.head()
```

```python
#df2 = df2.assign(year=df2['year'].str.replace('Stichtag.', ''))
df2['year'] = df2['year'].str.replace('Stichtag.', '', regex=True)
```

```python
df2['year'] = pd.to_datetime(df2['year'], errors='coerce', infer_datetime_format=True).dt.year.astype(int)
```
